Now let's try to use a new approach of feature engineering. That is use the existing columns to compute a third column. That is of a good use to us.

Take a look at the below dataset.

```python
import pandas as pd
```

```python
df_weights_heights = pd.read_csv('https://gist.githubusercontent.com/nstokoe/7d4717e96c21b8ad04ec91f361b000cb/raw/bf95a2e30fceb9f2ae990eac8379fc7d844a0196/weight-height.csv')
```

```python
df_weights_heights.head()
```

It seems like the height is in **inches** and the weight is in **pounds**. 

Now let's try to create a column that holds the BMI of each record.



*   BMI	=	body mass index
*   m	=	mass (in kilograms)
*   h	=	height (in meters)

Since the height should be in meters, let's convert the height column from inch to meters

```python
df_weights_heights['Height'] = df_weights_heights['Height'] *0.0254 # convert height from inches to meters
```

And since the weight is in pounds let's change it to kgs.

```python
df_weights_heights['Weight'] = df_weights_heights['Weight'] *0.453592 # convert weight from pounds(lbs) to kgs
```

### Interaction Feature
Next, we need to compute the BMI = mass/height^2. Since we are using a  combination of two or more features, which can be a product, sum, or difference, then we are engineering an interaction feature.

```python
df_weights_heights['BMI'] = (df_weights_heights['Weight'])/(df_weights_heights['Height']**2) # BMI = weight/(height^2)
```

```python
df_weights_heights.head()
```

### Indicator Feature 
Now that we have a new meaningful feature BMI. Let's create an indicator feature that tells us whether the person is overweight or not based on a threshold.


```python
df_weights_heights['Obesity Indicator'] = (df_weights_heights['BMI'] >= 25)*1 # True * 1 = 1, False * 1 = 0
```
