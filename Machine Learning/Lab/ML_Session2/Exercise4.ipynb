```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.fillna(0)
```

```python
df.head()
```

```python
df = pd.get_dummies(df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

Finding optimal Kmeans model

```python
from sklearn.cluster import KMeans
ssd = []

for k in range(2,30):
    model = KMeans(n_clusters=k)
    model.fit(scaled_df)
    ssd.append(model.inertia_)
plt.plot(range(2,30),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

Choosing the optimal K is up to the student

```python
best_model = KMeans(n_clusters = 4)
best_model.fit(scaled_df)

from sklearn.metrics import silhouette_score
print(silhouette_score(scaled_df, best_model.predict(scaled_df)))
```

Finding optimal Hierarchal Clustering model

```python
from sklearn.cluster import AgglomerativeClustering

num_cluster= np.arange(2,20)
scores=[]
for k in num_cluster:
    model = AgglomerativeClustering(n_clusters=k)
    cluster_labels = model.fit_predict(scaled_df)
    scores.append(silhouette_score(scaled_df, cluster_labels))
    
print(scores)
```

Choosing the optimal K is up to the student

```python
best_model = AgglomerativeClustering(n_clusters = 3)
best_model.fit(scaled_df)

from sklearn.metrics import silhouette_score
print(silhouette_score(scaled_df, best_model.predict(scaled_df)))
```

Finding optimal DBSCAN model 

```python
from sklearn.cluster import DBSCAN

outlier_percent = []

for eps in np.linspace(0.001,3,50):
    dbscan = DBSCAN(eps=eps)
    cluster_labels = dbscan.fit_predict(scaled_df)
    perc_outliers = 100 * np.sum(dbscan.labels_ == -1) / len(cluster_labels)
    outlier_percent.append(perc_outliers)


sns.lineplot(x=np.linspace(0.001,3,50),y=outlier_percent)
plt.ylabel("Percentage of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
```

Choosing the value of esp is up to the student
