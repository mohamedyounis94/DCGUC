## K-Means

**K-means** ==> finding cluster centers to group data points based on minimizing the sum of squared errors between each datapoint and its cluster center.  
K-means is very simple clustering algorithms.

**limitations**
*  depends on the starting point of the clusters
*  the number of clusters need to be defined beforehand.

```python
### import Libraries
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
angle = np.linspace(0,2*np.pi,20, endpoint = False)
X = np.append([np.cos(angle)],[np.sin(angle)],0).transpose()
```

```python
X= pd.DataFrame(X)
X.head()
sns.scatterplot(data=X)
```

```python
kmodel = KMeans(n_clusters=5)
kmodel.fit(X)
labels = kmodel.labels_
print(labels)
print("The error is {}".format(kmodel.inertia_))
```

Write code that calculates the inertia for 1 to 10 clusters, and plot the inertia as a function of the number of clusters.

```python
num_cluster = np.arange(1,11)
error = []
print(num_cluster)
for num in num_cluster:
  model = KMeans(n_clusters=num)
  model.fit(X)
  error.append(model.interia_)
```

```python
### BEGIN SOLUTION
inertia = []
list_num_clusters = list(range(1,11))
for num_clusters in list_num_clusters:
    km = KMeans(n_clusters=num_clusters)
    km.fit(X)
    inertia.append(km.inertia_)
print(inertia)    
plt.plot(list_num_clusters,inertia)
plt.scatter(list_num_clusters,inertia)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia');
### END SOLUTION
```
